```python
# 1. load vdj dataset
import pandas as pd
import numpy as np
vdjdb = pd.read_csv("../processed_data/vdjdb_preprocessd_nonull.txt", sep = "\t")
vdjdb.head()
```

```python
data=vdjdb[['gene','cdr3','v.segm','j.segm','mhc.a','mhc.b']]
target=vdjdb['antigen.epitope']
# 3. feature engineering

# 3.1 feature extraction

# 1) Dictionary feature extraction
from sklearn.feature_extraction import DictVectorizer
data=data.to_dict(orient="records") 
DicVec_transfer = DictVectorizer(sparse=False)
data = DicVec_transfer.fit_transform(data)
print("Dictionary feature extraction return:\n", data)
print("DictVectorized feature names：\n", DicVec_transfer.get_feature_names_out())
```

```python
# 2. split dataset
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data, target,
random_state=22)
print("x_train:\n", x_train.shape)
print("x_test:\n", x_test.shape)
```

```python
# 3) PCA
from sklearn.decomposition import PCA
PCA_transfer = PCA(n_components=0.9)
x_train_PCA = PCA_transfer.fit_transform(x_train)
x_test_PCA = PCA_transfer.transform(x_test)
print("PCA return:\n", x_train_PCA)
```

```python
# 4. Algorithm prediction

# 1) KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train, y_train)
y_predict = knn.predict(x_test)
print("predict test set：", y_predict)
print("accuracy：", knn.score(x_test, y_test))
```

```python
# Grid Search & cross validation
from sklearn.model_selection import GridSearchCV
knn = KNeighborsClassifier()
param = {"n_neighbors": [3, 5, 10]}
gc = GridSearchCV(knn, param_grid=param, cv=2)
gc.fit(x_train, y_train)
print("Accuracy：", gc.score(x_test, y_test))
# 训练验证集的结果
print("Best results in cross validation：", gc.best_score_)
print("k value chosen by gc：", gc.best_estimator_)
print("cross validation results：", gc.cv_results_)
```

```python
# 2)Naive Bayes
from sklearn.naive_bayes import MultinomialNB
mlb = MultinomialNB(alpha=1.0)
mlb.fit(x_train, y_train)
y_predict = mlb.predict(x_test)
print("Prediction：", y_predict)
print("Real category：", y_test)
print("Accuracy：", mlb.score(x_test, y_test))
```

```python
# 3) Decision Tree
from sklearn.tree import DecisionTreeClassifier
dc = DecisionTreeClassifier(max_depth=5,criterion='gini')
dc.fit(x_train, y_train)
print("Accuracy：", dc.score(x_test, y_test))
```

```python
# 4) Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
param = {"n_estimators": [120,200,300,500,800,1200], "max_depth": [5, 8, 15, 25,
30]}
gc = GridSearchCV(rf, param_grid=param, cv=2)
gc.fit(x_train, y_train)
print("Accuracy：", gc.score(x_test, y_test))
```
