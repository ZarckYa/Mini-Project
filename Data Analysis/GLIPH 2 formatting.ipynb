```python
#importing relevant libraries
import pandas as pd
import numpy as np
```

```python
#importing the vdjdjb_full.txt and converting it to a df so that I can format the data easily
vdjdb_full = pd.read_csv("vdjdb_full.txt", sep = "\t", low_memory=False)
display(vdjdb_full)
```

```python
#cutting down the vdjdb full txt file into a df that only uses specific columns needed for GLIPH 2
vdjdb_gliph = vdjdb_full[['cdr3.beta', 'v.beta', 'j.beta', 'cdr3.alpha']]
```

```python
#viewing a sample of the data to see its format
vdjdb_gliph.head()
```

```python
#renaming the columns to the format that GLIPH 2 uses.
vdjdb_gliph.rename(columns = {'cdr3.beta': 'CDR3b', 'v.beta': 'Vb', 'j.beta': 'Jb', 'cdr3.alpha': 'CRD3a' })

#adding a frequency column set to 1 for every row
vdjdb_gliph.loc[:, 'Frequency'] = 1
```

```python
#exporting as a csv file to upload to GLIPH 2
vdjdb_gliph.to_csv('vdjdb_gliph.csv', index=False)
```
