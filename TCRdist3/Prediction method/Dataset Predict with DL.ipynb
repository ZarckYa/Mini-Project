```python
import pandas as pd
import numpy as np
import tensorflow as tf
import tcellmatch.api as tm
import os
from sklearn.metrics import accuracy_score
```

```python
VDJdb_TRA_path = "D:\Master in UoB\TB2 of UoB\Mini Project\Mini-Project Development\TCRdist3\VDJdb_TRA.tsv"
VDJdb_TRA = pd.read_table("VDJdb_TRA.tsv")
display(VDJdb_TRA)

VDJdb_TRA_Train = VDJdb_TRA.sample(frac=0.8)
#display(VDJdb_TRA_Train)
VDJdb_TRA_Test = VDJdb_TRA.drop(VDJdb_TRA_Train.index)
#display(VDJdb_TRA_Test)
VDJdb_TRA_Train.to_csv("Train_Test_data/VDJdb_TRA_Train.tsv", sep='\t')
VDJdb_TRA_Test.to_csv("Train_Test_data/VDJdb_TRA_Test.tsv", sep='\t')
display(VDJdb_TRA_Test)

VDJdb_TRA_Train_path = "Train_Test_data/VDJdb_TRA_Train.tsv"
VDJdb_TRA_Test_path = "Train_Test_data/VDJdb_TRA_Test.tsv"
```

```python
TCR_Predict_Target = list(set(VDJdb_TRA_Test["Epitope"].tolist()))
print(type(TCR_Predict_Target))
print(TCR_Predict_Target)


# TCR_Predict_Target = ["KPVPEVKIL", "SIIAYTMSL", "FMIDVQQWG", "NLVPMVATV", 
#                       "AIKLDDKDP", "MVNTVAGAMK", "KLFIRQEEV", "EENLLDFVRF", "TTDPSFLGRY"]
# EBV = VDJdb_TRA[VDJdb_TRA["Epitope species"] == "EBV"]
# EBV_epitope = set(EBV["Epitope"].tolist())
# print(EBV_epitope)
```

```python
ffn = tm.models.EstimatorBinary()
```

```python
ffn.read_vdjdb_as_categorical_model(
    fns=VDJdb_TRA_Train_path,
    fn_blosum=None,
    antigen_ids=TCR_Predict_Target,
    blosum_encoding=False,
    is_train=True,
    chains="tra"
)
```

```python
print("Shape of (CDR3,antigen) sequences: ",ffn.x_train.shape)
# print("The head of TCR sequences: ",ffn.x_train[0])
print("Shape of covariates: ",ffn.covariates_train.shape)
# print("The head of covariates: ",ffn.covariates_train[0:5])
print("Shape of target set: ",ffn.y_train.shape)
# print("The head of target set: ",ffn.y_train[0:5])
```

```python
print(ffn.x_train[0])
```

```python
print(np.unique(ffn.y_train))
print(ffn.y_train[50])
```

```python
ffn.assign_clonotype(flavor="manhatten")
```

```python
#max_obs: Maximum number of observations per clonotype.
ffn.downsample_clonotype(max_obs=10)
```

```python
print("Shape of (CDR3,antigen) sequences: ",ffn.x_train.shape)
```

```python
#Clear test set.
ffn.clear_test_data()
ffn.read_vdjdb_as_categorical_model(
    fns=VDJdb_TRA_Test_path,
    antigen_ids=TCR_Predict_Target,
    fn_blosum=None,
    blosum_encoding=False,
    is_train=False,
    chains="tra"
)
#ffn.remove_overlapping_antigens(data="test")
#Assign clonotype by Manhatten distance.
ffn.assign_clonotype(flavor="manhatten", data="test")
#Downsample clonotypes to data stored in x_test.
ffn.downsample_clonotype(max_obs=10, data="test")
# Sample negative binding pairs for training.
#ffn.sample_negative_data(is_train=False)
```

```python
print("Shape of (CDR3,antigen) sequences for training: ",ffn.x_train.shape)
print("Shape of target set for training: ",ffn.y_train.shape)
print("Shape of (CDR3,antigen) sequences for test: ",ffn.x_test.shape)
print("Shape of target set for test: ",ffn.y_test.shape)
```

```python
print(np.unique(ffn.y_train))
y_test_oh = np.argmax(ffn.y_test, axis=1)
print(y_test_oh[0:100])
```

```python
#ffn.downsample_data(n=15000, data="train")
#ffn.downsample_data(n=5000, data="test")
```

```python
print("Shape of TCR CDR3 sequences for training: ",ffn.x_train.shape)
print("Shape of TCR CDR3 sequences for test: ",ffn.x_test.shape)
```

```python
ffn.build_self_attention(
    residual_connection=True,
    aa_embedding_dim=0,
    attention_size=[6, 6],
    attention_heads=[4, 4],
    optimizer='adam',
    lr=0.001,
    loss='bce',
    label_smoothing=0
)

```

```python
 ffn.train(
    epochs=1,
    max_steps_per_epoch=1,
    batch_size=5
    )
```

```python
ffn.evaluate()
print(ffn.evaluations)
ffn.predict()
```

```python
predict = np.argmax(ffn.predictions, axis=1)
print(predict[0:100])
```

```python
acc = accuracy_score(y_test_oh, predict)
print(acc)
```
