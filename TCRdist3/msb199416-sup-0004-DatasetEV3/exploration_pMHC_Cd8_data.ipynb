```python
def read_tcrs(fn):
    cell_table = pandas.read_csv(datadir+fn, sep=",", header=0)
    # Extract TCR sequences:
    raw_tcrs = cell_table["cell_clono_cdr3_aa"].values
    bound_state = cell_table[binder_labels]
    tcr_seqs = [[y.split(':') for y in x.split(';')] for x in raw_tcrs]
    tras_cell = [[y[1] for y in x if y[0] == 'TRA'] for x in tcr_seqs]
    trbs_cell = [[y[1] for y in x if y[0] == 'TRB'] for x in tcr_seqs]
    n_tras = np.array([len(x) for x in tras_cell])
    n_trbs = np.array([len(x) for x in trbs_cell])
    # Discard doublets
    print("Number of cells with doublets %i" % len(tras_cell))
    is_no_doublet = np.logical_and(
        n_tras <= 1,
        n_trbs <= 1
    )
    raw_tcrs_nd = [x for i, x in enumerate(raw_tcrs) if is_no_doublet[i] == True]
    tras_cell_nd = [x for i, x in enumerate(tras_cell) if is_no_doublet[i] == True]
    trbs_cell_nd = [x for i, x in enumerate(trbs_cell) if is_no_doublet[i] == True]
    bound_state_nd = bound_state.iloc[np.where(is_no_doublet)[0], :]
    print("Number of cells without putative doublets %i" % len(tras_cell_nd))
    return raw_tcrs_nd, tras_cell_nd, trbs_cell_nd, bound_state_nd
```

```python
def read_clonotypes(fn):
    clonotype_table = pandas.read_csv(datadir+fn, sep=",", header=0)
    # Extract TCR sequences:
    raw_tcrs_clono = clonotype_table["cdr3s_aa"].values
    tcr_seqs_clono = [[y.split(':') for y in x.split(';')] for x in raw_tcrs_clono]
    tras_clono = [[y[1] for y in x if y[0] == 'TRA'] for x in tcr_seqs_clono]
    trbs_clono = [[y[1] for y in x if y[0] == 'TRB'] for x in tcr_seqs_clono]
    return raw_tcrs_clono, tras_clono, trbs_clono
```

## Sequence diversity

```python
from typing import List
eos_char = '_'
aa_list = [
    'A', 'R', 'N', 'D', 'C',
    'Q', 'E', 'G', 'H', 'I',
    'L', 'K', 'M', 'F', 'P',
    'S', 'T', 'W', 'Y', 'V',
    'B', 'J', 'Z', 'X', '*',
    eos_char
]
dict_aa = dict(zip(aa_list, range(len(aa_list))))

def read_blosum(fn):
    blosum_tab = pandas.read_csv(
        filepath_or_buffer=fn,
        skiprows=2, header=0,
        delimiter=" ", skipinitialspace=True,
        index_col=0
    )
    return blosum_tab.to_dict(orient="list")


def encode_as_blosum(
        x: List[str],
        fn_blosum: str
) -> np.ndarray:
    blosum_embedding = read_blosum(fn_blosum)
    dim_obs = len(x)
    dim_pos = np.max([len(xi) for xi in x])
    dim_aa = len(next(iter(blosum_embedding.values())))

    x_encoded = np.zeros([dim_obs, dim_pos, dim_aa])
    for i, xi in enumerate(x):  # Loop over observations.
        for k, aa in enumerate(xi):  # Loop over observed sequence positions.
            x_encoded[i, k, :] = blosum_embedding[aa]
        # Fill remaining positions as None.
        for k in np.arange(len(xi), dim_pos):  # Loop over padded remaining sequence positions.
            x_encoded[i, k, :] = 1
    return x_encoded

def encode_as_onehot(
        x: List[str],
):
    dim_obs = len(x)
    dim_pos = np.max([len(xi) for xi in x])
    dim_aa = len(aa_list)

    x_encoded = np.zeros([dim_obs, dim_pos, dim_aa])
    for i, xi in enumerate(x):  # Loop over observations.
        for k, aa in enumerate(xi):  # Loop over observed sequence positions.
            x_encoded[i, k, dict_aa[aa]] = 1
        # Fill remaining positions as None.
        for k in np.arange(len(xi), dim_pos):  # Loop over padded remaining sequence positions.
            x_encoded[i, k, dict_aa[eos_char]] = 1
    return x_encoded
```

Plot pairwise distance distribution of TCRs of a given specificity, both in one-hot encoded and in blosum50 space. We downsample each clonotype to a maximum of 10 observations to reduce skew of the pairwise distance distribution by large clonotypes. The total number of pairwise distances is very large, here we look at a random sample of 10,000 pairwise distances per specificity.

```python
nsamples = 1000
n_clonotype_max = 10
dists_onehot = np.zeros([2, 4, 51, nsamples])
dists_blosum = np.zeros([2, 4, 51, nsamples])

bound_states = [bound_state_1, bound_state_2, bound_state_3, bound_state_4]
for chain in range(2):  # TRA and TRB
    if chain == 0:
        seqs_cell = [tras_cell_1, tras_cell_2, tras_cell_3, tras_cell_4]
        seqs_clonotypes = [tras_clono_1, tras_clono_2, tras_clono_3, tras_clono_4]
    else:
        seqs_cell = [trbs_cell_1, trbs_cell_2, trbs_cell_3, trbs_cell_4]
        seqs_clonotypes = [trbs_clono_1, trbs_clono_2, trbs_clono_3, trbs_clono_4]
    for i in range(4):  # donors
        seqs_cell_i = seqs_cell[i]
        seqs_clonotypes_i = seqs_clonotypes[i]
        bound_state_i = bound_states[i]

        idx_included = np.where(np.logical_and(
            np.array([len(x) > 0 for x in seqs_cell_i]),
            np.sum(bound_state_i.values, axis=1) > 0
        ))[0]
        seqs_i_onehot = encode_as_onehot([seqs_cell_i[ii][0] for ii in idx_included])
        seqs_i_blosum = encode_as_blosum([seqs_cell_i[ii][0] for ii in idx_included], fn_blosum)

        # Downsample clonotypes:
        clonotypes = np.unique([x[0] for x in seqs_clonotypes_i if len(x) > 0])
        clonotype_count = np.zeros([len(clonotypes)])
        idx_to_keep = []
        idx_kept = []
        # Iterate over data set and check whether clonotypes are matched
        exclusion_counter = 0
        for iii, ii in enumerate(idx_included):
            # Assign clonotype
            temp = np.where(seqs_cell_i[ii][0] == clonotypes)[0][0]
            clonotype_count[temp] += 1
            if clonotype_count[temp] <= 10:
                idx_kept.append(iii)
            else:
                exclusion_counter += 1
        print("donor %i, chain %s: excluded %i/%i cells from large clonotypes" % 
              (i+1, ["TRA", "TRB"][chain], exclusion_counter, len(idx_included)))
        idx_included = idx_included[idx_kept]
        seqs_i_onehot = seqs_i_onehot[idx_kept]
        seqs_i_blosum = seqs_i_blosum[idx_kept]

        labels = [x.split("_")[1] for x in bound_state_i.columns] + ["none"]
        bound_states_i_idx = np.argmax(bound_state_i.values[idx_included, :], axis=1)

        for j, label in enumerate(labels):
            idx_label = np.where(bound_states_i_idx == j)[0]
            if len(idx_label) > 10:
                pair_idx = np.concatenate([
                    np.expand_dims(np.random.choice(idx_label, size=2, replace=False),axis=1) 
                    for _ in range(nsamples)
                ], axis=1)
                dists_onehot[chain, i, j, :] = np.sum(
                    np.sum(np.abs(seqs_i_onehot[pair_idx[0]] - seqs_i_onehot[pair_idx[1]]), axis=2) > 0, 
                    axis=-1
                )
                dists_blosum[chain, i, j, :] = np.sqrt(np.sum(np.sum(np.square(
                    seqs_i_blosum[pair_idx[0]] - seqs_i_blosum[pair_idx[1]]
                ), axis=-1), axis=-1))
```

```python
# Labels that pass frequency filter
idx_passing_onehot = np.where(np.sum(np.sum(np.sum(dists_onehot, axis=0), axis=0), axis=-1) > 0)[0]
fig, axs = plt.subplots(
    ncols=dists_onehot.shape[1],
    nrows=len(idx_passing_onehot), 
    figsize=(20, 4*len(idx_passing_onehot))
)
for i in range(dists_onehot.shape[1]):
    for ij, j in enumerate(idx_passing_onehot):
        if np.any(dists_onehot[:, i, j, :] != 0):
            sns.distplot(
                dists_onehot[0, i, j, :], 
                kde=False,
                ax=axs[ij, i],
                label='TRA'
            )
            sns.distplot(
                dists_onehot[1, i, j, :], 
                kde=False,
                ax=axs[ij, i],
                label='TRA'
            )
            if ij == 0:
                axs[ij, i].set_title('donor ' + str(i+1))
            else:
                axs[ij, i].set_title('')
            if i == 0 and ij == 0:
                axs[ij, i].legend()
        if i == 0:
            axs[ij, i].set_ylabel(labels[ij])
        if ij == len(idx_passing_onehot)-1:
            axs[ij, i].set_xlabel("manhatten distance")
plt.legend(prop={'size': 12})
plt.tight_layout()
plt.savefig(figir+'supp_x_a.pdf')
plt.show()
```

### Data set size if clonotypes are downsampled

```python
def downsample_clono(assignments, max_per_clono):
    clonotypes = np.unique(assignments)
    n_cells = assignments.shape[0]
    idx_to_keep = np.concatenate([np.random.choice(
        a=np.arange(0, n_cells)[assignments == x],
        size=np.min([max_per_clono, np.sum(assignments == x)]),
        replace=False
    ) for x in np.unique(assignments)], axis=0)
    print("data set size downsampled to 10 cells per clonotypes reduces from %i to %i" % (len(assignments), len(idx_to_keep)))
    return len(assignments), len(idx_to_keep)
```
